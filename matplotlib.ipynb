```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd


np.random.seed(0)
data = np.random.randn(100)
categories = ['A', 'B', 'C', 'D']
category_counts = [10, 15, 7, 12]


plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
plt.plot(data, marker='o', color='b', linestyle='-')
plt.title('Line Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)


plt.subplot(2, 2, 2)
plt.scatter(np.arange(100), data, color='r', marker='x')
plt.title('Scatter Plot')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)

# 3. Bar Plot
plt.subplot(2, 2, 3)
sns.barplot(x=categories, y=category_counts, palette='viridis')
plt.title('Bar Plot')
plt.ylabel('Counts')

# 4. Histogram
plt.subplot(2, 2, 4)
plt.hist(data, bins=15, color='g', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# 5. Pie Chart
plt.figure(figsize=(8, 6))
plt.pie(category_counts, labels=categories, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Pie Chart')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular
plt.show()

# 6. Count Plot (Using Seaborn)
# Creating a DataFrame for count plot
df = pd.DataFrame({'Category': np.random.choice(categories, size=100)})
plt.figure(figsize=(8, 6))
sns.countplot(x='Category', data=df, palette='husl')
plt.title('Count Plot')
plt.xlabel('Category')
plt.ylabel('Counts')
plt.show()






```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample Data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Set the style of seaborn
sns.set(style="whitegrid")

# Create a figure and a set of subplots
fig, ax = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1: Basic Line Plot
ax[0, 0].plot(x, y, marker='o', color='b', label='Sine Wave')
ax[0, 0].set_title('Basic Line Plot')
ax[0, 0].set_xlabel('X-axis')
ax[0, 0].set_ylabel('Y-axis')
ax[0, 0].legend()
ax[0, 0].grid(True)

# Plot 2: Scatter Plot
y2 = np.random.rand(100)
ax[0, 1].scatter(x, y2, marker='x', color='r', label='Random Scatter')
ax[0, 1].set_title('Scatter Plot')
ax[0, 1].set_xlabel('X-axis')
ax[0, 1].set_ylabel('Y-axis')
ax[0, 1].legend()
ax[0, 1].grid(True)

# Plot 3: Histogram
data = np.random.randn(1000)
ax[1, 0].hist(data, bins=30, color='purple', alpha=0.7)
ax[1, 0].set_title('Histogram')
ax[1, 0].set_xlabel('Value')
ax[1, 0].set_ylabel('Frequency')

# Plot 4: Box Plot
sns.boxplot(x=data, ax=ax[1, 1], color='green')
ax[1, 1].set_title('Box Plot')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
